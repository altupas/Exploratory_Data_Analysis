**Part 1: Exploratory Data Analysis**


**1. Using the mpg dataset, identify:**
1.   Response variable:  hwy (highway miles per gallon), This variable represents the fuel efficiency of a vehicle when driving on highways. It is the outcome of interest because we want to understand how fuel efficiency changes based on engine characteristics.

2.   Explanatory variable: displ (engine displacement in liters), This variable represents the size of the vehicle’s engine. It is used as the explanatory variable because the engine size is expected to influence fuel efficiency.



**2. Create the following visualizations using the mpg dataset:**

*  A scatterplot of hwy vs displ

```r
ggplot(mpg, aes(x = displ, y = hwy)) +
  geom_point(alpha = 0.6) +
  labs(
    title = "Scatterplot of Highway MPG vs Engine Displacement",
    x = "Engine Displacement (liters)",
    y = "Highway Miles per Gallon (hwy)"
  ) +
  theme_minimal()
```

**Interpretation:**

The plot show negative relationship as engine displacement decreases the highway fuel effeciency also increases.

*   One additional plot (histogram or boxplot) for either hwy or displ

```r
ggplot(mpg, aes(x = hwy)) +
  geom_histogram(bins = 15) +
  labs(
    title = "Histogram of Highway Miles per Gallon",
    x = "Highway Miles per Gallon (hwy)",
    y = "Frequency"
  ) +
  theme_minimal()
```

**Part 2: Building a Simple Linear Model**

1. Fit a simple linear regression model using the mpg dataset with:

*   hwy as the response variable
*   displ as the explanatory variable

```r
model <- lm(hwy ~ displ, data = mpg)

plot(hwy ~ displ, data = mpg,
     main = "Linear Regression: Highway MPG vs Engine Displacement",
     xlab = "Engine Displacement (liters)",
     ylab = "Highway MPG")

abline(model, col = "red", lwd = 2)

summary(model)
```

2. Write the fitted model equation and interpret:

*   The slope in terms of engine displacement and highway MPG
*   The intercept in context of the data

For every 1 liter increase in engine displacement, the expected highway fuel efficiency decreases by 3.53 miles per gallon. The model predicts that a car with an engine displacement of 0 liters would have a highway fuel efficiency of 35.7 mpg."

Note for context: Since a car cannot have a 0-liter engine, this intercept is theoretical and acts mostly as a baseline for the mathematics of the mode.

**Part 3: Model Assumptions and Diagnostics**

1.  **Using your fitted mpg linear model, assess the following assumptions using diagnostic plots:**

```r
library(ggplot2)
mpg_lm <- lm(hwy ~ displ, data = mpg)

summary(mpg_lm)
```

*  Linearity & Constant variance (homoscedasticity)

```r
plot(mpg_lm, which = 1)
```

*   Normality of residuals

```r
plot(mpg_lm, which = 2)
```

**2. Briefly explain whether each assumption appears to be satisfied for the mpg model.**

**Linearity**

The Residuals vs Fitted plot shows a generally random scatter around zero, indicating that a linear relationship between engine displacement and highway MPG is reasonable. While slight curvature may be present, the linearity assumption is mostly satisfied.

**Normality of Residuals**

The Normal Q–Q plot shows that most residuals lie close to the reference line, with minor deviations at the tails. This suggests that the residuals are approximately normally distributed.

**Constant Variance (Homoscedasticity)**

The spread of residuals appears relatively consistent across fitted values, although there is slight variation at higher engine displacements. Overall, the constant variance assumption is reasonably satisfied.

**Part 4: Transformation or Model Critique**

```r
log_model <- lm(log(hwy) ~ log(displ), data = mpg)

summary(log_model)

par(mfrow = c(2, 2))
plot(log_model)
```

**1. If any assumptions are violated in the mpg model, apply an appropriate transformation (e.g., log transformation of displ or hwy) OR justify why no transformation is necessary.**

Transformation Justification The diagnostic plots in Part 3 revealed that the Linearity assumption was violated. Because the plot showed a distinct curved "U-shape" pattern, indicating that a straight line does not capture the relationship between engine size and fuel efficiency well. We then applied a Log-Log transformation to linearizes power-law relationships, which are common in physical data like engine mechanics.

**2. Refit the model if transformed and briefly compare it with the original mpg model, or discuss the limitations of the simple linear model for this dataset.**

I refitted the model using the equation:
$\ln(\text{hwy}) = \beta_0 + \beta_1 \ln(\text{displ})$.

Original Model: Had a lower $R^2$ (approx 0.58) and biased residuals.

Transformed Model: The $R^2$ improved (approx 0.72), indicating the log model explains more of the variance in fuel efficiency. The residuals in the new model appear much more random and centered around zero, suggesting the Log-Log model provides a better fit for the data.
